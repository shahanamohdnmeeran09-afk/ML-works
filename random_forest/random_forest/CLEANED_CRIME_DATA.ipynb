```python
import matplotlib.pyplot as plt  
import pandas as pd  
import numpy as np 
import seaborn as sns
```

```python
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")
data
```

```python
data.info()
```

## Drop Useless Columns

```python
data.isnull().sum()
```

```python
cols_to_drop = ["Mocodes ","Vict Sex ","Vict Descent","Premis Desc","Weapon Used Cd","Weapon Desc","Crm Cd 2","Crm Cd 3","Crm Cd 4","Cross Street"]
data = data.drop(columns=[col for col in cols_to_drop if col in data.columns])
```

```python
data
```

## Fix Missing Values in Remaining Columns

```python
data['Premis Cd'] = data['Premis Cd'].fillna(data['Premis Cd'].mode()[0])
```

```python
data['Premis Cd'].isnull().sum()
```

```python
data['Status'] = data['Status'].fillna(data['Status'].mode()[0])
```

```python
data['Crm Cd 1'] = data['Crm Cd 1'].fillna(data['Crm Cd 1'].mode()[0])
```

```python
# Make Sure LAT/LON Are Numeric
data['LAT'] = pd.to_numeric(data['LAT'], errors='coerce')
data['LON'] = pd.to_numeric(data['LON'], errors='coerce')

data = data.dropna(subset=["LAT", "LON"])
```

```python
data['LAT']
```

```python
sns.boxplot(x=data['LAT'])
plt.title("Outliers - Latitude")
plt.show()

sns.boxplot(x=data['LON'])
plt.title("Outliers - Longitude")
plt.show()

```

```python
data = data[
    (data['LAT'].between(33, 35)) &
    (data['LON'].between(-118.8, -117.5))
]
```

```python
sns.boxplot(x=data['LAT'])
plt.title("Outliers - Latitude")
plt.show()

sns.boxplot(x=data['LON'])
plt.title("Outliers - Longitude")
plt.show()
```

```python
#Convert Dates & Time
data['DATE OCC'] = pd.to_datetime(data['DATE OCC'])

data['Year'] = data['DATE OCC'].dt.year
data['Month'] = data['DATE OCC'].dt.month
data['Day'] = data['DATE OCC'].dt.day
data['Weekday'] = data['DATE OCC'].dt.weekday   # 0 = Monday, 6 = Sunday

```

```python
# Convert time to hour 
data['Hour'] = data['TIME OCC'] // 100
```

## Encode Categorical Columns

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data['Crm Cd Desc'] = le.fit_transform(data['Crm Cd Desc'])
data['Status'] = le.fit_transform(data['Status'])

```
